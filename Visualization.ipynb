<h1> Visualizing the Population in Lebnaon accross Districs</h1>   


```python
import geopandas as gpd
import matplotlib.pyplot as plt

#Shape File from https://data.humdata.org/dataset/lebanon-administrative-boundaries-levels-0-3
districts_df = gpd.read_file('Lebanon Shape File/lbn_admbnda_adm2_cdr_20200810.shp')

districts_df.head()
```

```python
#Selecting the desired columns from the original dataset
districts_new_df = districts_df[['admin2Name', 'geometry']].copy()

#Creating a Dictionary with Districts existed in the shape file as Keys and 
# the respective population gotten from https://www.citypopulation.de/en/lebanon/admin/
population = {
    'Akkar':423596 ,
    'El Hermel':69997,
    'El Minieh-Dennie':206075,
    'Tripoli':342760,
    'Zgharta':81490,
    'Baalbek':387935,
    'El Koura':75056,
    'El Batroun':57339,
    'Bcharre':28231,
    'Jbeil':99388,
    'Kesrwane':182834,
    'El Meten':523593,
    'Zahle':336382,
    'Beirut':433249,
    'Baabda':568714,
    'Aley':196282,
    'West Bekaa':150838,
    'Chouf':231427,
    'Rachaya':47122,
    'Jezzine':31575,
    'Saida':287987,
    'El Nabatieh':155233,
    'Hasbaya':37784,
    'Marjaayoun':85960,
    'Sour':270516,
    'Bent Jbeil':104862
}

# adding a new column to the dataset using the dictionary 
districts_new_df['Population'] = districts_new_df['admin2Name'].map(population)

districts_new_df
```

```python
# figure style and size
plt.style.use('seaborn-white')
fig,ax = plt.subplots(figsize=(25,22))

# plotting and allocating colors using cmap using the documentation from https://matplotlib.org/2.0.2/users/colormaps.html
districts_new_df.plot(ax=ax, edgecolor='black', column='Population', legend=True, cmap='coolwarm')

# annoting each polygon aka. each district
districts_new_df.apply(lambda x: ax.annotate(text=x.admin2Name, xy=x.geometry.centroid.coords[0], ha='center', fontsize=19, weight='bold'),axis=1);

#setting the title
ax.set_title('Population by Districts in Lebanon in 2017-12-31', size=33)

#getting the overall population by summing cumulatively the population column and getting the last row's value
overall_population = districts_new_df['Population'].cumsum(axis=0).tail(1).values[0]

# annotating the overall population
ax.annotate('Overall Population: {}'.format(overall_population),xycoords='figure fraction', xy=(0.4,0.1), size=27)

#x and y labels font size
fig.axes[0].tick_params(labelsize=25)

#bar legend font size 
fig.axes[1].tick_params(labelsize=25)
```
